```python
model=tensorflow.keras.models.load_model('/content/drive/MyDrive/PFE/Classification model/Classifier.h5')
```

```python
from keras.preprocessing import image
testImage='/content/drive/MyDrive/PFE/Classification model/Test Images/img5.jpg'
test_image=image.load_img(testImage,target_size=(128, 128))
test_image=image.img_to_array(test_image)
test_image=np.expand_dims(test_image,axis=0)
result=model.predict(test_image,verbose=0)
```

```python
img=cv2.imread(testImage)
plt.imshow(img)
plt.title('picture')
plt.show()

if np.argmax(result)==0:
  qlty = 'Fresh'
else:
  qlty = 'Rotten'

print('Quality is: ',qlty)
```
